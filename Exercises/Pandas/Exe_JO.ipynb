```python
# 22. Quels sont les joueurs ayant gagné des médailles d’Or en badminton simple au cours des années. Trier le résultat par ordre alphabétique sur le nom des joueurs.

gold_badminton_singles = data[(data['Sport'] == 'Badminton') & (data['Event'].str.contains('Singles')) & (data['Medal'] == 'Gold')]

gold_badminton_singles.sort_values('Name')
```

```python
# 23. Quels sont les 3 pays ayant gagné le plus de médailles de 1984 à 2016

medals = data[(data['Medal'] == 'Gold') | (data['Medal'] == 'Silver') | (data['Medal'] == 'Bronze')]
period = medals[(medals['Year'] > 1984) & (medals['Year'] < 2016)]

countries = period['Team'].value_counts()

countries.head(3)
```
