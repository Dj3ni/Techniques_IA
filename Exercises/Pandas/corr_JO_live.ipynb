23) Quels sont les 3 pays ayant gagné le plus de médailles de 1984 à 2016

```python
medal = df['Medal'].notna()
year84_16 = df['Year'].between(1984,2016)

df[medal&year84_16].groupby(by=['NOC']).count()['Medal']\
    .sort_values(ascending=False)
```

```python
df[medal&year84_16]\
    .drop_duplicates(subset=['NOC','Year','Event','Medal','Team'])\
        .groupby(by=['NOC']).count()['Medal']\
            .sort_values(ascending=False)
```

30) Quel est le top 15 des pays ayant le plus de médailles (toutes médailles confondues)

```python
df[medal].drop_duplicates(subset=['NOC','Year','Event','Medal','Team'])\
    .groupby(by=['NOC']).count()[['Medal']]\
        .sort_values(by=['Medal'],ascending=False).head(15)
```

31) Quels sont les pays gagnant les plus médaillés par sport distinct 

```python
df.loc[(df["Medal"].isin(["Gold", "Silver", "Bronze"]))]\
                           .drop_duplicates(subset=["Event","Medal","Year"])\
                           .groupby(by=['Sport','NOC']).count()[['Medal']]\
                           .reset_index()\
                           .sort_values(by=['Sport','Medal'],ascending=[True,False])\
                           .drop_duplicates(subset=['Sport'])
```

32) Quels étaient les pays les plus efficaces durant la guerre froide de 1952 à 1988 (i.e. le pays 
avec le plus de médailles par nombre d’athlètes envoyés)

```python
cold_war = df['Year'].between(1952,1988)

efficiency = df[cold_war].groupby(by = ['NOC']).count()[['ID','Medal']]

efficiency['eff'] = (efficiency['Medal'] / efficiency['ID']) * 100

efficiency.sort_values(by=['eff'],ascending=False)
```
